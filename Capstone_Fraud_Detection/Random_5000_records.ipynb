```python
# Example: a DataFrame with an integer index of IDs 1..100000
n = 100000000
# df = pd.DataFrame({
#     "value": np.arange(n) * 2  # some dummy data
# }, index=np.arange(1, n + 1))  # index is 1..100000

# Generate 5000 random integers in [1, 100000]
rng = np.random.default_rng(seed=51)  # optional seed for reproducibility
rand_ids = rng.integers(1, n + 1, size=600000)

# (Optional) Make them unique to avoid repeated rows; keeps order of first occurrence
# If you prefer to retain duplicates, skip the next line.
rand_ids_unique = np.unique(rand_ids)

# Keep only IDs that actually exist in the DataFrame index
valid_ids = rand_ids_unique[np.isin(rand_ids_unique, df.index)]

# Filter rows by label (index) using .loc
filtered = df.loc[valid_ids]

# print(filtered.shape)
# print(filtered.head())
filtered.to_csv("Test_Load_5000.csv")
```
